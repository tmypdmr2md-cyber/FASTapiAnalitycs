```python
import requests
```

```python
path1 = "http://localhost:8002/"
path2 = "api/events/"

response = requests.post(f"{path1}{path2}", json={"path": "/test+", "description": "test event"})


if response:
    if response.status_code == 200:
        print("ok")
        print("Status:", response.status_code)
    data = response.json()
    print(data)
else: 
    print("Error:", response.status_code)
```

```python
import requests
from uuid import UUID, uuid4
path1 = "http://localhost:8002/"
id = uuid4()
path2 = "api/events/"
endpoint = f"{path1}{path2}{id}" 
# без path2 мы обращаемся к корневым функциям main.py, а с path2 - к функциям, которые мы описали в routing.py
endpoint
```

```python
data = {"path" : "/test/path/", "description": "test description"}
```

```python
headers = {
    "Content-Type": "application/json"
    }

response = requests.put(endpoint, headers=headers, json=data)
if response.ok:
    print("ok")
    print("Status:", response.status_code)
    data = response.json()
    print(data)
else: 
    print("Error:", response.status_code)
```
