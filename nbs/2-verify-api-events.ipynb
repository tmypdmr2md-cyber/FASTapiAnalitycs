```python
import requests

path = "/api/events/12"
base_url = "http://localhost:8002"
endpoint = f"{base_url}{path}"
endpoint
```

```python
response = requests.get(endpoint)
```

```python
if response.ok:
    print("Status:", response.status_code)
    print("OK", response.ok)
    data = response.json()
    print(f"{data}")
else:
    print("Failed:", response.status_code)
    print("OK", response.ok)
```
