```python
import requests as req
```

```python
path_healthz = "/healthz"
path_events_router = "/api/events/"
base_url = "http://localhost:8002"

endpoint_healthz = f"{base_url}{path_healthz}"
endpoint_events = f"{base_url}{path_events_router}"

endpoint_healthz, endpoint_events
```

```python
response = req.get(endpoint_healthz)

assert response.ok, f"healthz failed: {response.status_code}"
data = response.json()
assert data == {"status": "ok"}, f"unexpected payload: {data}"
print("healthz OK:", data)
```

```python
response = req.get(endpoint_events)

assert response.ok, f"events failed: {response.status_code}"
data = response.json()
assert "items" in data, f"missing items key: {data}"
print("events OK:", data)
```
